```python
import pandas as pd
import numpy as np

train = pd.read_csv('D:/pandas_main/part7/extracted_files/open/train.csv')

# ì •ìˆ˜í˜• ë³€í™˜ í•¨ìˆ˜
import re
def extraced_and_convert_to_int(data):
    numbers = re.findall(r'\d+',str(data))
    return int(numbers[0]) if numbers else np.nan

train['ê°€í•´ìš´ì „ì ì—°ë ¹'] = train['ê°€í•´ìš´ì „ì ì—°ë ¹'].apply(extraced_and_convert_to_int)
train['í”¼í•´ìš´ì „ì ì—°ë ¹'] = train['í”¼í•´ìš´ì „ì ì—°ë ¹'].apply(extraced_and_convert_to_int)
```

```python

test = pd.read_csv('D:/pandas_main/part7/extracted_files/open/test.csv')
train = pd.read_csv('D:/pandas_main/part7/extracted_files/open/train.csv')


time_pattern = r'(\d{4})-(\d{1,2})-(\d{1,2}) (\d{1,2})'

train[['ë…„','ì›”','ì¼','ì‹œê°„']] =  train['ì‚¬ê³ ì¼ì‹œ'].str.extract(time_pattern).apply(pd.to_numeric)
train_df = train.drop(columns=['ì‚¬ê³ ì¼ì‹œ'])

test[['ë…„','ì›”','ì¼','ì‹œê°„']] =  test['ì‚¬ê³ ì¼ì‹œ'].str.extract(time_pattern).apply(pd.to_numeric)
test_df = test.drop(columns=['ì‚¬ê³ ì¼ì‹œ'])
```

```python
# ìœ„ì¹˜ íŠ¹ì„± ì¶”ì¶œ : ECLO ê°€ ë†’ì€ ì‹œêµ°êµ¬ ê°€ ìˆë‹¤ëŠ”ê²ƒì„ í™•ì¸
local_pattern = r'(\S+) (\S+) (\S+)'  # ì„œìš¸ì‹œ ê°•ë‚¨êµ¬ ëŒ€ì¹˜ë™
train_df[['ë„ì‹œ','êµ¬','ë™']] = train['ì‹œêµ°êµ¬'].str.extract(local_pattern)
train_df = train_df.drop(columns=['ì‹œêµ°êµ¬']) 

test_df[['ë„ì‹œ','êµ¬','ë™']] = test['ì‹œêµ°êµ¬'].str.extract(local_pattern)
test_df = test_df.drop(columns=['ì‹œêµ°êµ¬']) 
```

```python
road_pattern = r'(.+) - (.+)'

train_df[['ë„ë¡œí˜•íƒœ1','ë„ë¡œí˜•íƒœ2']] = train_df['ë„ë¡œí˜•íƒœ'].str.extract(road_pattern)
train_df = train_df.drop(columns=['ë„ë¡œí˜•íƒœ'])

test_df[['ë„ë¡œí˜•íƒœ1','ë„ë¡œí˜•íƒœ2']] = test_df['ë„ë¡œí˜•íƒœ'].str.extract(road_pattern)
test_df = test_df.drop(columns=['ë„ë¡œí˜•íƒœ'])

train_df['ê°€í•´ìš´ì „ì ì—°ë ¹'].head()
```

```python
pattern = r'(\d+)'
train_df['ê°€í•´ìš´ì „ì ì—°ë ¹'] = train_df['ê°€í•´ìš´ì „ì ì—°ë ¹'].str.extract(pattern).astype(float)
train_df['í”¼í•´ìš´ì „ì ì—°ë ¹'] = train_df['í”¼í•´ìš´ì „ì ì—°ë ¹'].str.extract(pattern).astype(float)

train_df
```

```python
# ì§€ì—­ë³„ í‰ê·  ì—°ë ¹ ì¶”ì¶œ ë° ë°ì´í„° ë³‘í•©
# ê°€í•´ìš´ì „ì ì—°ë ¹ ê³¼ í”¼í•´ìš´ì „ì ì—°ë ¹ì˜ í‰ê· ì„ ì§€ì—­ë³„ë¡œ ê³„ì‚°  -> ì›ë³¸ ë°ì´í„°ì™€ ë³‘í•©
age_mean = train_df[['ë„ì‹œ','êµ¬','ë™','ê°€í•´ìš´ì „ì ì—°ë ¹','í”¼í•´ìš´ì „ì ì—°ë ¹']].groupby(['ë„ì‹œ','êµ¬','ë™']).mean()
age_mean.columns = ['ê°€í•´ìš´ì „ì í‰ê· ì—°ë ¹', 'í”¼í•´ìš´ì „ì í‰ê· ì—°ë ¹']
train_df = pd.merge(train_df, age_mean, how='left', on = ['ë„ì‹œ','êµ¬','ë™'])
test_df = pd.merge(test_df, age_mean, how='left', on = ['ë„ì‹œ','êµ¬','ë™'])
train_df['ê°€í•´ìš´ì „ì í‰ê· ì—°ë ¹'].head()
```

```python
# ì„±ë³„ ë‚¨:1  ì—¬:0
train_df['ê°€í•´ìš´ì „ì ì„±ë³„'] = train_df['ê°€í•´ìš´ì „ì ì„±ë³„'].map({'ë‚¨':1 , 'ì—¬': 0})
train_df['í”¼í•´ìš´ì „ì ì„±ë³„'] = train_df['í”¼í•´ìš´ì „ì ì„±ë³„'].map({'ë‚¨':1 , 'ì—¬': 0})
```

```python
# ì§€ì—­ë³„ í‰ê·  ì„±ë³„

gender_mean = train_df[['ë„ì‹œ','êµ¬','ë™','ê°€í•´ìš´ì „ì ì„±ë³„','í”¼í•´ìš´ì „ì ì„±ë³„']].groupby(['ë„ì‹œ','êµ¬','ë™']).mean()
gender_mean.columns = ['ê°€í•´ìš´ì „ì í‰ê· ì„±ë³„','í”¼í•´ìš´ì „ì í‰ê· ì„±ë³„']
train_df = pd.merge(train_df,gender_mean,how='left', on=['ë„ì‹œ','êµ¬','ë™'])
test_df = pd.merge(test_df,gender_mean,how='left', on=['ë„ì‹œ','êµ¬','ë™'])
```

```python
# ì£¼ë§ íŠ¹ì„± í†  , ì¼  1  ë‚˜ë¨¸ì§€ëŠ” 0
train_df['ì£¼ë§'] = train_df['ìš”ì¼'].apply(lambda x : 1 if x in ['í† ìš”ì¼','ì¼ìš”ì¼'] else 0)
test_df['ì£¼ë§'] = test_df['ìš”ì¼'].apply(lambda x : 1 if x in ['í† ìš”ì¼','ì¼ìš”ì¼'] else 0)
```

```python
# ì§€ì—­ë³„ ì‚¬ê³  ìœ í˜•ì— ë”°ë¥¸ ì°¨ëŸ‰ ì¢…ë¥˜ì˜ ìœ„í—˜ë„ ì‚°ì¶œ ë° ë°ì´í„° ë³‘í•©

# ì‚¬ê³  ìœ í˜•ê³¼ ì°¨ì¢…ë³„ ì‚¬ê³  ë¹ˆë„ëŠ” ì§€ì—­ë³„ë¡œ ê°ê¸° ë‹¤ë¥¸ íŠ¹ì„±ì„ ê°€ì§„ë‹¤. - ECLOì˜ ì£¼ìš” ì›ì¸

# íŒŒìƒë³€ìˆ˜ : ê°€í•´ìš´ì „ì ì°¨ì¢…ì˜ ECLO ì™€ ê°€í•´ ìš´ì „ì ì°¨ì¢… ê±´ìˆ˜ë¥¼ ê³±í•´ì„œ ì°¨ì¢…ë³„ ìœ„í—˜ë„ë¥¼ ë§Œë“ ë‹¤ -> íŒŒìƒë³€ìˆ˜

ride_dangerous = train_df[['ê°€í•´ìš´ì „ì ì°¨ì¢…','ECLO']].groupby('ê°€í•´ìš´ì „ì ì°¨ì¢…').mean()
ride_dangerous.columns = ['ride_dangerous']

ride_count = train_df[['ë„ì‹œ','êµ¬','ê°€í•´ìš´ì „ì ì°¨ì¢…']]
ride_count['count'] = 1
ride_count = ride_count.groupby(['ë„ì‹œ','êµ¬','ê°€í•´ìš´ì „ì ì°¨ì¢…']).count().reset_index()
temp_1 = pd.merge(ride_count , ride_dangerous, how='left', on=['ê°€í•´ìš´ì „ì ì°¨ì¢…'])
temp_1['multiply'] = temp_1['count'] * temp_1['ride_dangerous']
temp_1 = temp_1.groupby(['ë„ì‹œ','êµ¬']).sum().reset_index().drop(columns=['ride_dangerous'])
temp_1['ride_dangerous'] = temp_1['multiply'] / temp_1['count']
temp_1 = temp_1.drop(columns=['ê°€í•´ìš´ì „ì ì°¨ì¢…',	'count',	'multiply'])
temp_1

train_df = pd.merge(train_df, temp_1, how='left',on=['ë„ì‹œ','êµ¬'])
test_df = pd.merge(test_df, temp_1, how='left',on=['ë„ì‹œ','êµ¬'])
test_df
```
